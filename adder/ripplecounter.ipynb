```python
import jkflipflop
import clock
import notgate
```

```python
antimony = notgate.antimony + jkflipflop.antimony + '''
model ripplecounter(C, Q0, Q1)
    jk1: jkflipflop(J, C, K, Q0, _Qbar1)
    ng: notgate(Q0, notQ0)
    jk2: jkflipflop(J, notQ0, K, Q1, _Qbar2)
    
    J = 1;
    K = 1;
end
'''
```

```python
testmodel = antimony + clock.antimony + '''
model testmodel()
    c: clock(a, p, ps)
    rc: ripplecounter(S, Q0, Q1)
    S := c.signal
    a = 1; p = 300; ps = 0;
end
'''
def test():

    import tellurium as te
    period = 300
    end = period * 10 # end time
    r = te.loada(testmodel)
    r.p = period
    
    sim = r.simulate(0, end, end, ['time', 'Q0', 'Q1', 'c_signal'])
    te.plotWithLegend(r, sim)
    
    cycles = xrange(0, end, period)
    
    cycle_values = [[0, 0], [1, 0], [0, 1], [1, 1]];
    threshold = 0.2
    
    for i, c in enumerate(cycles):
        if i == 0:
            continue
        expected = cycle_values[(i + 2)%4]
        t = c + period;
        q0 = sim[t][1]
        q1 = sim[t][2]
        resid_q0 = abs(expected[0] - q0)
        resid_q1 = abs(expected[1] - q1)
        if resid_q0 > threshold:
            print 'Q0: %f out of range at t: %d. Expected Q0: %d, Q1: %d' % (q0, t, expected[0], expected[1])
        if resid_q1 > threshold:
            print 'Q1: %f out of range at t: %d. Expected Q0: %d, Q1: %d' % (q1, t, expected[0], expected[1])
        
    return sim
sim = test()
```
