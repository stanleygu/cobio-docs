Importing the half adder and OR gate modules.

```python
halfadder = __import__('half-adder')
import orgate
```

Now I wire up the full adder.

```python
fulladder = orgate.antimony + halfadder.antimony + '''
model fulladder(A, B, Cin, S, Cout)
    ha1: halfadder(A, B, S1, C1);
    ha2: halfadder(S1, Cin, S, C2);
    org: orgate(C1, C2, Cout);
end

'''
```

I use this test function to see a sample response from the adder.

```python
def test():
    
    testmodel = fulladder + '''
    model test
        a: fulladder(A, B, Cin, S, Cout);

        #at (time < 100): A = 0, B = 0, Cin = 0 // S=0, Cout=0
        at (time > 100): A = 1, B = 0, Cin = 0 // S=1, Cout=0
        at (time > 200): A = 0, B = 1, Cin = 0 // S=1, Cout=0
        at (time > 300): A = 1, B = 1, Cin = 0 // S=0, Cout=1
        at (time > 400): A = 0, B = 0, Cin = 1 // S=1, Cout=0
        at (time > 500): A = 1, B = 0, Cin = 1 // S=0, Cout=1
        at (time > 600): A = 0, B = 1, Cin = 1 // S=0, Cout=1
        at (time > 700): A = 1, B = 1, Cin = 1 // S=1, Cout=1

    end
    '''

    import tellurium as te
    r = te.loada(testmodel)
    sim = r.simulate(0, 800, 800, ['time', 'S', 'Cout'])
    te.plotWithLegend(r, sim)
```

```python
sim = test()
```
