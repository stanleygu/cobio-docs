```python
import notebooktools as nb
nb.loadNotebooksAsModules()
```

```python
import jkflipflop
import clock
import notgate
```

```python
antimony = notgate.antimony + clock.antimony + jkflipflop.antimony + '''
model counter()
    c: clock(a, p, ps)
    jk1: jkflipflop(J, c.signal, K, Q0, _Qbar1)
    ng: notgate(Q0, notQ0)
    jk2: jkflipflop(J, notQ0, K, Q1, _Qbar2)
    
    J = 1;
    K = 1;
    
    a = 1; p = 100; ps = 0;
end
'''
```

```python
import tellurium as te
r = te.loada(antimony)
r.simulate(0, 200)
sim = r.simulate(0, 1000, 1000, ['time', 'Q0', 'Q1'])
te.plotWithLegend(r, sim)
```
