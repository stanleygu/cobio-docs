```python
from tellurium.visualization import SBMLDiagram
import BIOMD0000000051 as m
```

```python
diagram = SBMLDiagram(m.sbmlString, reactions={
    'shape': 'box'
}, reactants={
    'arrowhead': 'none'
}, modifiers={
    'arrowhead': 'odot'
})
for r in diagram.model.reactions:
    n = diagram.G.get_node(r.getId())
    n.attr['label'] = ''
    n.attr['width'] = 0.1
    n.attr['height'] = 0.1
    
diagram.draw(layout='dot')
```

```python
for r in diagram.model.reactions:
    n = diagram.G.get_node(r.getId())
    n.label = ''
```
